# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "../Resources/city_data.csv"
ride_data_to_load = "../Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type

urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]

urban_ride_count = urban_cities_df.groupby(["city"]).count()["ride_id"]
suburban_ride_count = suburban_cities_df.groupby(["city"]).count()["ride_id"]
rural_ride_count = rural_cities_df.groupby(["city"]).count()["ride_id"]
```

```python
# 2. Get the total drivers for each city type

urban_cities1_df = city_data_df[city_data_df["type"] == "Urban"]
suburban_cities1_df = city_data_df[city_data_df["type"] == "Suburban"]
rural_cities1_df = city_data_df[city_data_df["type"] == "Rural"]


urban_driver_count = urban_cities1_df.groupby(["city"]).sum()["driver_count"]
suburban_driver_count = suburban_cities1_df.groupby(["city"]).sum()["driver_count"]
rural_driver_count = rural_cities1_df.groupby(["city"]).sum()["driver_count"]
```

```python
#  3. Get the total amount of fares for each city type

urban_fares = urban_cities_df.groupby(["city"]).sum()["fare"]
suburban_fares = suburban_cities_df.groupby(["city"]).sum()["fare"]
rural_fares = rural_cities_df.groupby(["city"]).sum()["fare"]
```

```python
#  4. Get the average fare per ride for each city type. 

urban_avg_ride_fare = urban_cities_df.groupby(["type"]).sum()["fare"]/urban_ride_count.sum()
suburban_avg_ride_fare = suburban_cities_df.groupby(["type"]).sum()["fare"]/suburban_ride_count.sum()
rural_avg_ride_fare = rural_cities_df.groupby(["type"]).sum()["fare"]/rural_ride_count.sum()
```

```python
# 5. Get the average fare per driver for each city type. 

urban_avg_driver_fare = urban_cities_df.groupby(["type"]).sum()["fare"]/urban_driver_count.sum()
suburban_avg_driver_fare = suburban_cities_df.groupby(["type"]).sum()["fare"]/suburban_driver_count.sum()
rural_avg_driver_fare = rural_cities_df.groupby(["type"]).sum()["fare"]/rural_driver_count.sum()

```

```python
#  6. Create a PyBer summary DataFrame. 

city_type = ["Rural", "Suburban", "Urban"]
total_rides = [rural_ride_count.sum(), suburban_ride_count.sum(), urban_ride_count.sum()]
total_drivers = [rural_driver_count.sum(), suburban_driver_count.sum(), urban_driver_count.sum()]
total_fares = [rural_fares.sum(), suburban_fares.sum(), urban_fares.sum()]
avg_ride_fares = [float(rural_avg_ride_fare), float(suburban_avg_ride_fare), float(urban_avg_ride_fare)]
avg_driver_fare = [float(rural_avg_driver_fare), float(suburban_avg_driver_fare), float(urban_avg_driver_fare)]

pyber_summary_df = pd.DataFrame(
        {"type": city_type,
         "Total Rides": total_rides,
         "Total Drivers": total_drivers,
         "Total Fares": total_fares,
         "Average Fare per Ride": avg_ride_fares,
         "Average Fare per Driver": avg_driver_fare})

pyber_summary_df = pyber_summary_df.set_index('type', drop=True)

pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.

pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)

pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)

pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame

pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

type_sum_fares_df = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
type_sum_fares_df.tail(10)
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.

type_sum_fares_df = type_sum_fares_df.reset_index()
type_sum_fares_df.tail(10)
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

type_sum_fares_pivot = type_sum_fares_df.pivot(index="date", columns="type", values="fare")
type_sum_fares_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

given_dates_sum_fares_df = type_sum_fares_pivot.loc['2019-01-01':'2019-04-29']
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.

given_dates_sum_fares_df.index = pd.to_datetime(given_dates_sum_fares_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()

given_dates_sum_fares_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.

sum_fares_week = given_dates_sum_fares_df.resample("W").sum()
sum_fares_week.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

sum_fares_week.plot.line(figsize=(15,6))
plt.title("Total Fare by City Type")
plt.ylabel("Fare ($USD)")
plt.xlabel("")
plt.legend(loc="center")
plt.savefig("PyBer_fare_summary.png")
```
